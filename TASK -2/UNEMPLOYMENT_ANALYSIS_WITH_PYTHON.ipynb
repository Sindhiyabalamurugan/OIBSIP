**UNEMPLOYMENT ANALYSIS WITH PYTHON**

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Dataset

```python
data = pd.read_csv('/content/Unemployment in India.csv')
data.head()
```

EDA

```python
data.info()
```

```python
print(data.columns)
```

```python
# Remove leading and trailing spaces from column names
data.columns = data.columns.str.strip()

# Convert the 'Date' column to a datetime object
data['Date'] = pd.to_datetime(data['Date'])
```

```python

data.describe()
```

```python
#Check for missing values
missing_count = data.isna().sum()
print("Missing value \n",missing_count)
```

```python
#Data distribution and histograms
plt.figure(figsize=(12,6))
sns.histplot(data['Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of umeployment Rate')
plt.xlabel('Umemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the average unemployment rate
average_unemployment_rate = data['Estimated Unemployment Rate (%)'].mean()

# Print the result
print(f'Average Unemployment Rate: {average_unemployment_rate}')
```

```python
# Convert the 'Date' column to a datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Create a line plot for both unemployment rate and employed rate over time
plt.figure(figsize=(10, 6))

# Plot the unemployment rate
plt.plot(data['Date'], data['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='violet', label='Unemployment Rate')

# Plot the employed rate
plt.plot(data['Date'], data['Estimated Employed'], marker='o', linestyle='-', color='g', label='Employed Rate')

plt.title('Unemployment Rate and Employed Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()  # Show the legend to differentiate the lines

# Show the plot
plt.show()
```

```python
# Create a line plot for unemployment rate over time
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='violet')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Show the plot
plt.show()
```
